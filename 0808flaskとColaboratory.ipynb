```python
!pip install flask pyngrok
import os
from flask import Flask
from pyngrok import ngrok, conf

# ngrokトークンを設定
conf.get_default().auth_token = "取得したトークンを記入"

app = Flask(__name__)

@app.route("/")
def home():
    return "Hello, World!"

if __name__ == "__main__":
    public_url = ngrok.connect(5000)
    print(f"ngrok URL: {public_url}")
    app.run(port=5000)
```
